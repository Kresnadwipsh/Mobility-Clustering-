## Data Cleaning

```python
df1.fillna(0,inplace = True)
df1.drop(df1[df1['Nama Kota'] == 'LUAR DKI JAKARTA'].index, inplace = True)
df1.drop(df1[df1['Nama Kota'] ==  'PROSES UPDATE DATA'].index, inplace = True)
df1.drop(['ID Kelurahan','Nama Provinsi'],axis = 1, inplace = True)
```

```python
df2['tanggal']= pd.to_datetime(df2['tanggal'])
df2 = df2[df2['tanggal'].between('2021-12-1','2022-1-31')]
```

```python
df3.fillna(0,inplace = True)
```

```python
df_waze = pd.concat([df4,df5],axis = 0)
df_waze = pd.merge(df_waze,df6,how ='left',on='street')
df_waze.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5'],axis = 1, inplace = True)
df_waze = df_waze.rename(columns={'date':'Tanggal'})
```
