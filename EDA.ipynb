## Exploratory Data Analysis

- Covid - 19  

```python
plt.figure(figsize=(30,9))
plt.style.use('seaborn')
ax1 = sns.lineplot(data= df_dki, x='Tanggal', y='POSITIF', color = 'red')
ax2 = sns.lineplot(data= df_dki, x='Tanggal', y='Positif Sembuh', color = 'green')
plt.title('Covid-19 DKI Jakarta Period : December 1, 2021 - January 31, 2022', size = 20)
plt.xlabel('Date',size= 15)
plt.ylabel('Total Cases',size= 15)
plt.axvline(18993, 0,5, color = 'black')
plt.legend(['Positive','Healed','New Year Celebration'])
plt.show()
```

```python
plt.figure(figsize=(30,9))
plt.style.use('seaborn')
ax1 = sns.lineplot(data= df_dki, x='Tanggal', y='POSITIF', color = 'red')
ax2 = sns.lineplot(data= df_jabar, x='tanggal', y='confirmation_total', color = 'b')
ax3 = sns.lineplot(data= df_jatim, x='tanggal', y='Konfirmasi (Sebelum)', color = 'orange')
plt.title('Comparison of Covid-19 Period: December 1, 2021 - January 31, 2022', size = 20)
plt.xlabel('Date',size= 15)
plt.ylabel('Total Cases',size= 15)
plt.axvline(18993, 0,5, color = 'black')
plt.legend(['DKI Jakarta','West Java','East Java','New Year Celebration'])
plt.show()
```

```python
rate = df_dki[['SUSPEK','PROBABLE','PELAKU PERJALANAN','KONTAK ERAT','DISCARDED','POSITIF','Positif Sembuh','Positif Meninggal']].corr()

plt.figure(figsize=(10,5))
sns.heatmap(rate, cmap='ocean_r', annot=True)
plt.show()
```

```python
covid = df_dki.groupby(['Nama Kecamatan'])[['POSITIF']].sum().sort_values(by='POSITIF',ascending=True).reset_index()

y = covid['POSITIF']
x = covid['Nama Kecamatan']

plt.figure(figsize = (30,12))
ax = plt.barh(x, y, color='red')
plt.ylabel('District')
plt.xlabel('Total Cases')
plt.title('Total Cases of Covid-19 in Each District',size= 20)
for i in ax.patches:
    plt.text(i.get_width()+0.02, i.get_y()+0.003,
             str(round((i.get_width()), 2)),
             fontsize=15,
             color='black')
plt.show()
```

- Traffic Mobility

```python
plt.figure(figsize=(30,9))
sns.distplot(df_waze['speed_kmh'])
plt.show()
```

```python
df_waze['Tanggal'] = pd.to_datetime(df_waze['Tanggal'])

plt.figure(figsize=(30,9))
sns.lineplot(data= df_waze, x='Tanggal', y='speed_kmh', color = 'steelblue')
plt.title('Average Speed ​​During 2022 New Year Celebration', size = 20)
plt.xlabel('Date',size= 15)
plt.ylabel('Average Speed Km/h',size= 15)
plt.axvline(18993, 0,5, color = 'black')
plt.legend(['Average Speed','New Year Celebration'])
plt.show()
```

```python
df_tb = df_waze[df_waze['Tanggal'].between('2021-12-19','2022-01-08')]
df_tb = df_tb.groupby(['Nama Kecamatan'])[['speed_kmh','length','delay']].mean().sort_values(by='speed_kmh',ascending=False).reset_index()

y = df_tb['speed_kmh']
x = df_tb['Nama Kecamatan']

plt.figure(figsize = (30,12))
plt.barh(x, y)
plt.ylabel('District')
plt.xlabel('Average Speed Km/h')
plt.title('Average Speed in Each District Periode: 19 Desember 2021 - 8 Januari 2022',size= 20)
plt.show()
```

```python
df_waze['Tanggal']= pd.to_datetime(df_waze['Tanggal'])
df_mob = pd.merge(df_dki,df_waze,how ='left',on=['Tanggal','Nama Kecamatan'])
mobility = df_mob[['POSITIF','speed_kmh','length','delay']].corr()

plt.figure(figsize=(10,5))
sns.heatmap(mobility, cmap='ocean_r', annot=True)
plt.show()
```

```python
scat = df_mob.groupby(['Nama Kecamatan'])[['speed_kmh','POSITIF','length','delay']].mean().reset_index()

f, axes = plt.subplots(1, 3,figsize=(20,5))
sns.regplot(x='POSITIF',y='speed_kmh', data=scat, color='b',ax=axes[0])
sns.regplot(x='POSITIF',y='length', data=scat, color='r', ax=axes[1])
sns.regplot(x='POSITIF',y='delay', data=scat, color='g', ax=axes[2])
sns.despine(bottom=True, left=True)
axes[0].set(xlabel='Positive', ylabel='Average Speed')
axes[1].set(xlabel='Positive', ylabel='Length')
axes[2].set(xlabel='Positive', ylabel='Delay')
axes[1].yaxis.set_label_position("left")
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(30,9))
ax1 = sns.lineplot(data= df_des, x='date', y='retail_and_recreation_percent_change_from_baseline', color = 'red')
ax2 = sns.lineplot(data= df_des, x='date', y='grocery_and_pharmacy_percent_change_from_baseline', color = 'b')
ax3 = sns.lineplot(data= df_des, x='date', y='parks_percent_change_from_baseline', color = 'g')
ax4 = sns.lineplot(data= df_des, x='date', y='transit_stations_percent_change_from_baseline', color = 'orange')
ax5 = sns.lineplot(data= df_des, x='date', y='workplaces_percent_change_from_baseline', color = 'purple')
ax6 = sns.lineplot(data= df_des, x='date', y='residential_percent_change_from_baseline', color = 'pink')
plt.title('Mobility by Destination', size = 20)
plt.xlabel('Date',size= 15)
plt.ylabel('Change',size= 15)
plt.axvline(13, 0,5, color = 'black')
plt.legend(['Retail and Recreation','Grocery and Farmacy','Park','Transit Station','Workplace','Residential','New Year Celebration'])
plt.show()
```
